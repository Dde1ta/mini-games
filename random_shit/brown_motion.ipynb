```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
time = 1
pos = [0, 0]
random_x = np.random.normal
random_y = np.random.normal
```

```python
def step(pos) -> [int]:
    pos[0] = pos[0] + math.sqrt(time) * random_x(0, 1)
    pos[1] = pos[1] + math.sqrt(time) * random_x(0, 1)
    return pos
```

```python
x = [0]
y = [0]
for i in range(100):
    pos = step(pos)
    x.append(i)
    y.append(pos[1])
```

```python
plt.plot(x, y)
plt.show()
```
